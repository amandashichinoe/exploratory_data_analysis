### De todos os biblionários, quantos exercem a posição de CEO?

```python
# Verificando os cargos dos bilionários do dataset
for title in df.title.values[:10]:
    print(title)
```

```python
# Criando uma lista onde, se title é CEO, retorna 1. Caso contrário, retorna 0
contains_ceo = np.array([0 if type(title) is not str else 1 if 'ceo' in title.lower() else 0 for title in df.title.values])
```

```python
print(f'{contains_ceo.mean()*100:.2f}% dos bilionários do dataset possuem cargos de CEO')
```

Considerou-se como sendo cargos de CEO todos os cargos cujo título possui a palavra "CEO", entretanto, 87,15% dos dados dessa coluna (title) são nulos e isso impacta negativamente a análise.
Esse percentual de CEOs foi calculado como sendo a média da lista 'contain_ceo', onde foi considerado como sendo '1', ou seja, "é um CEO", todos os títulos que continham CEO. Todo o restante (incluindo os valores nulos) foram considerados como 0.

Abaixo, é calculada a mesma porcentagem, porém excluindo os valores nulos.

```python
contains_ceo_not_null = np.array([-1 if type(title) is not str else 1 if 'ceo' in title.lower() else 0 for title in df.title.values])
```

```python
print(f'{contains_ceo_not_null[contains_ceo_not_null != -1].mean()*100:.2f}% dos bilionários do dataset, cujo o cargo está discriminado, possuem cargos de CEO')
```

A partir disso, tem-se que para todos os cargos que são não nulos, 27% dos bilionários desempenham o papel de CEO.

### Quais são as 10 indústrias com mais bilionários?

```python
# Verificando as 10 indústrias com mais bilionários.
# *a coluna rank foi usada para a contagem porque não possui valores nulos
top_industries = df.groupby('industries').count()[['rank']].reset_index().sort_values(by='rank', ascending=False).rename(columns={'rank':'quantity'}).reset_index(drop=True).head(10)
top_industries
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.barplot(x='quantity', y='industries', data=top_industries, palette="viridis")

# Configurações de eixos, título e estilos do gráfico
plt.title('Top 10 Indústrias com maior quantidade de Bilionários', fontsize=16)
plt.xlabel(None)
plt.ylabel(None)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.grid(False)

plt.show()
```

A partir do dataframe e gráfico acima, verifica-se as 10 indústrias com maior número de bilionários.

### Qual é a distribuição de gênero entre os bilionários?

```python
# Verificando a quantidade de biblionários por gênero
df['gender'].value_counts().reset_index().rename(columns={'count':'quantity'})
```

```python
# Plotando um gráfico da distribuição de gênero entre os bilionários do dataset

sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))

# Contabilizando a quantidade por gênero
gender_counts = df['gender'].value_counts()

# Criando o gráfico de barras
barplot = sns.barplot(x=gender_counts.index.map({'M': 'Homens', 'F': 'Mulheres'}), y=gender_counts.values, palette='viridis')

# Adicionando rótulos nas barras com as porcentagens
total_bilionarios = gender_counts.sum()
for p in barplot.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total_bilionarios)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    barplot.annotate(percentage, (x, y), ha='center', va='bottom')

# Definindo rótulos, título e estilo
plt.title('Quantidade de Bilionários por Gênero')
plt.xlabel(None)
plt.ylabel(None)

ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.grid(False)

# Exibindo o gráfico
plt.show()

```

O gráfico acima exibe a distribuição de bilionários por gênero. Dos bilionários presentes no dataset, a maioria são homens (87,2%), enquanto as mulheres representam apenas 12,8%.

Para plotar um gráfico de mapas utilizando Geopandas, é necessário importar as informações de coordenadas para cada país. Ao fazê-lo, verificou-se que alguns países estavam com nome diferente no dataset. Optou-se por renomeá-los no dataset em estudo.
O dataframe acima possui os países que não estão presentes no geopandas. Nessa análise eles não serão considerados no mapa de distribuição de bilionários pelo mundo.

```python
# Plotando mapa com a distribuição de bilionários por país

# Carregando o mapa
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

billionaires_by_country = df['country'].value_counts().reset_index()

# Unindo os DataFrames 'world' e 'billionaires_by_country'
world = world.merge(billionaires_by_country, how='left', left_on='name', right_on='country')

# Preenchendo os países sem dados com 0
world['count'] = world['count'].fillna(0)

# Criando um mapa
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

world.plot(column='count', ax=ax, cmap='viridis', edgecolor='0.8')

# Adicionando uma barra de cores e uma legenda
vmin, vmax = world['count'].min(), world['count'].max()
sm = plt.cm.ScalarMappable(cmap='viridis', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []  # truque para a barra de cores funcionar
cbar = fig.colorbar(sm, ax=ax, orientation='vertical', fraction=0.03, pad=0.1)
cbar.set_label('Quantidade de Bilionários')

# Adicionando o título
plt.title('Distribuição de Bilionários por País', fontsize=16)

plt.show()
```

```python
# Plotando um mapa da distribuição de riqueza dos bilionários por país
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

wealth_by_country = df.groupby('country')['finalWorth'].sum().reset_index()

world = world.merge(wealth_by_country, how='left', left_on='name', right_on='country')

world['finalWorth'] = world['finalWorth'].fillna(0)

fig, ax = plt.subplots(1, 1, figsize=(15, 10))

world.plot(column='finalWorth', ax=ax, cmap='viridis', edgecolor='0.8')

vmin, vmax = world['finalWorth'].min(), world['finalWorth'].max()
sm = plt.cm.ScalarMappable(cmap='viridis', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []  # truque para a barra de cores funcionar
cbar = fig.colorbar(sm, ax=ax, orientation='vertical', fraction=0.03, pad=0.1)
cbar.set_label('Total da Fortuna (em bilhões de dólares)')

plt.title('Distribuição da Riqueza dos Bilionários por País', fontsize=16)

plt.show()
```
